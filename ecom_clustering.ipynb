```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
import plotly.figure_factory as ff
import numpy as np
from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import linkage
from sklearn.preprocessing import StandardScaler
from scipy.cluster.hierarchy import linkage, dendrogram
```

```python
df=pd.read_csv('/content/ecom cust seg.csv')
df.head()
```

```python
# Fill missing values in 'Gender' and convert to numeric
df['Gender'].fillna(df['Gender'].mode()[0], inplace=True)
df['Gender'] = df['Gender'].map({'M': 1, 'F': 0})
```

```python
df.isnull().sum()
```

```python

# Assuming your 'Gender' column has values 0 for Female and 1 for Male
gender_counts = df['Gender'].value_counts()

fig = px.pie(
    gender_counts,
    values=gender_counts.values,
    names=['Female', 'Male'],  # Provide labels for the pie slices
    title="Gender Distribution"
)
fig.update_layout(template="plotly_dark")
fig.show()
```

```python
columns_for_pie = ['Jordan', 'Gatorade', 'Samsung', 'Asus', 'Udis', 'Mondelez International', 'Wrangler', 'Vans', 'Fila', 'Brooks', 'H&M', 'Dairy Queen', 'Fendi', 'Hewlett Packard', 'Pladis', 'Asics', 'Siemens', 'J.M. Smucker', 'Pop Chips', 'Juniper', 'Huawei', 'Compaq', 'IBM', 'Burberry', 'Mi', 'LG', 'Dior', 'Scabal', 'Tommy Hilfiger', 'Hollister', 'Forever 21', 'Colavita', 'Microsoft', 'Jiffy mix', 'Kraft']

# Calculate the sum of values for each specified column
column_sums = df[columns_for_pie].sum()

fig = px.pie(
    values=column_sums.values,
    names=column_sums.index,
    title="Distribution of Values Across Columns"
)
fig.update_layout(template="plotly_dark")
fig.show()
```

```python
# Normalize the data
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df)
```

```python
# Determine Optimal K using Elbow Method ----
inertia = []
k_values = range(1, 11)

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(df_scaled)
    inertia.append(kmeans.inertia_)

```

```python
# Plot Elbow
fig_elbow = go.Figure()
fig_elbow.add_trace(go.Scatter(x=list(k_values), y=inertia, mode='lines+markers', marker=dict(color='blue')))
fig_elbow.update_layout(title="Elbow Method for Optimal K",
                        xaxis_title="Number of Clusters (K)",
                        yaxis_title="Inertia",
                        template="plotly_dark")
fig_elbow.show()
```

```python
#  Apply KMeans Clustering
optimal_k = 5
kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
df['Cluster_KMeans'] = kmeans.fit_predict(df_scaled)

```

```python
fig_kmeans = px.scatter_3d(df, x='Orders', y='Samsung', z='Microsoft',
                           color=df['Cluster_KMeans'].astype(str),
                           title='KMeans Clustering', labels={'color': 'Cluster'})
fig_kmeans.update_layout(template="plotly_dark")
fig_kmeans.show()
```

```python

fig_2 = px.scatter_3d(df, x='Orders', y='Burberry', z='Asus',
                           color=df['Cluster_KMeans'].astype(str),
                           title='KMeans Clustering', labels={'color': 'Cluster'})
fig_2.update_layout(template="plotly_dark")
fig_2.show()
```

```python

# Apply Hierarchical Clustering
hierarchical = AgglomerativeClustering(n_clusters=optimal_k)
df['Cluster_Hierarchical'] = hierarchical.fit_predict(df_scaled)
```

```python
#Visualize with Dendrogram (using a subset of data)
import plotly.figure_factory as ff
from scipy.cluster.hierarchy import linkage, dendrogram


n_samples_for_dendrogram = 100
df_subset = df_scaled[:n_samples_for_dendrogram]

# Create linkage matrix
linkage_matrix = linkage(df_subset, method='ward')  # You can experiment with different linkage methods

# Create dendrogram
fig_dendrogram = ff.create_dendrogram(linkage_matrix, color_threshold=20) # Adjust color_threshold as needed

# Update layout for better visualization
fig_dendrogram.update_layout(
    title='Hierarchical Clustering Dendrogram',
    width=800,
    height=500,
    xaxis_title="Data Points",
    yaxis_title="Distance"
)
fig_dendrogram.update_layout(template="plotly_dark")
fig_dendrogram.show()
```
